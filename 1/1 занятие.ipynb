сортировка пузырьком по возрастанию:
a1....an
итеративно:
будем сравнивать рядом стоящие переменные, и если ai > a(i+1), будем их менять местами. После одного такого цикла самым последним эелементом будет максимальный. Если такой цикл сделать n раз, то последовательность отсоритруется
O(n^2)

```python
def bubbleSort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-1, i, -1):
            if arr[j] < arr[j-1]:
                arr[j], arr[j-1] = arr[j-1], arr[j]
    return arr

#O(n * n)
#верхний алгоримт будет работать n раз, второй цикл будет работать n-1 раз
#операции с if имеет сложность О(1)
bubbleSort([2,3,1,0])
```
