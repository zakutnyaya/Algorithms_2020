Алгоритмы - это эффективность кода по операциям (как количество операций растет в зависимости от количества эелементов, котрые нужно обработать) и количество места, которое используется во время работы алгоритмов.
 О-нотация: верхний предел работы алгоритма.  То есть самом худшем случае не будет больше такого-то количества операций.
 
 Омега - это нижний предел работы алгоритмы.
 
 O(n) - линейный поиск - сколько элементов, столько и операций (напечатать элементы, проверить на четность и тд)
 
 Нужно стремиться к логарифмическому времени O(logn). Это очень быстрый алгоритм. Когда на каждом шаге мы сокращаем операции вдвое.

```python
# бинарный поиск: работает на отсортированных списках и работает быстрее, чем линейный поиск
# список делим на 2 и сравниваем середину с нужным числом
lst = [2, 5, 78, 190, 412, 942, 2412, 2522532]

#942
def binary_search(a, value):
    mid = len(a) // 2
    low = 0
    high = len(a) - 1
    
    while a[mid] != value and low <= high:
        if value > a[mid]:
            low = mid+1
        else:
            high = mid - 1
        mid = (low + high) // 2
            
    if low > high:
        print("No value")
    else:
        print('ID = ', mid)
        
binary_search(lst, 2412)
```
